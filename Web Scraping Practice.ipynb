# Web Scraping via lxml

```python
import pandas as pd
```

```python
import requests
```

```python
import lxml
```

```python
from lxml import etree
```

```python
from lxml import html
```

```python
url = "http://www.whatmobile.com.pk"
```

```python
path = '//*[@id="container"]/div[2]/div[2]'
```

```python
response = requests.get(url)
```

```python
byte_data = response.content
```

```python
source_code = html.fromstring(byte_data) 
```

```python
tree = source_code.xpath(path) 
```

```python
print(tree[0].text_content())
```
